### <font color="blue">Introduction to Linear Regression Analysis

Linear regression is an algorithm that provides a linear relationship between an independent variable and a dependent variable to predict the outcome of future events. It is a statistical method used in data science and machine learning for predictive analysis.

The independent variable is also the predictor or explanatory variable that remains unchanged due to the change in other variables. However, the dependent variable changes with fluctuations in the independent variable. The regression model predicts the value of the dependent variable, which is the response or outcome variable being analyzed or studied.

Thus, linear regression is a supervised learning algorithm that simulates a mathematical relationship between variables and makes predictions for continuous or numeric variables such as sales, salary, age, product price, etc.

This analysis method is advantageous when at least two variables are available in the data, as observed in stock market forecasting, portfolio management, scientific analysis, etc.

#### 1. Easy implementation

The linear regression model is computationally simple to implement as it does not demand a lot of engineering overheads, neither before the model launch nor during its maintenance.

#### 2. Interpretability

#### 3. Scalability

Linear regression is not computationally heavy and, therefore, fits well in cases where scaling is essential. For example, the model can scale well regarding increased data volume (big data). 

#### 4. Optimal for online settings

The ease of computation of these algorithms allows them to be used in online settings. The model can be trained and retrained with each new example to generate predictions in real-time, unlike the neural networks or support vector machines that are computationally heavy and require plenty of computing resources and substantial waiting time to retrain on a new dataset. ### 1. Key Assumption 

Linear regression analysis can be practical and be performed accurately when the data abides by a set of rules. These are referred to as key assumptions concerning the data.

#### 1.1. Linear relationship 
The first important assumption of linear regression is that the dependent and independent variables should be linearly related. The relationship can be determined with the help of scatter plots that help in visualization. Also, one needs to check for outliers as linear regression is sensitive to them.

#### 1.2. Normal distribution of residuals
The second assumption relates to the normal distribution of residuals or error terms, i.e., if residuals are non-normally distributed, the model-based estimation may become too wide or narrow. The non-normal distribution also underscores that you need to closely observe some unusual data points to make a good model.

#### 1.3. Multicollinearity
The third assumption relates to multicollinearity, where several independent variables in a model are highly correlated. More correlated variables make it difficult to determine which variable contributes to predicting the target variable. Also, standard errors inevitably increase due to correlated variables.

Moreover, with such a robust variable correlation, the predicted regression coefficient of a correlated variable further depends on the other variables available in the model, leading to wrong conclusions and poor performance. The goal, therefore, is to have minimal or lesser multicollinearity.

#### 1.4. Autocorrelation
One fundamental assumption of linear regression specifies that the given dataset should not be autocorrelated. This mostly happens when residuals or error terms are not independent of each other.

In other words, the situation arises when the value of f(a+1) is not independent of the value of f(a). For example, in the case of stock prices, the price of one stock depends on the cost of the previous one.

#### 1.5. Homoscedasticity
Another assumption of linear regression analysis is referred to as homoscedasticity. Homoscedasticity relates to cases where the residuals (error terms) between the independent and dependent variables remain the same for all independent variable values.

In simple words, the residuals or error terms must have constant variance. If not, it leads to an unbalanced scatter of residuals, known as heteroscedasticity. With heteroscedasticity, you cannot trust the results of the regression analysis.

### 2. Use Simple Model

The common misconception is that complex problems require complex regression models. However, research has identified that simpler models offer precise predictions as they effectively show how well the data fit with the models.

This has its implications, as the more complex the model is, the more tailored the model will be to the specific dataset. As a result, generalizability suffers.

One must verify, validate, and ensure that the added complexity produces narrower prediction intervals. Also, keep checking the predicted R-squared value rather than chasing the high R-squared range. R-squared is a statistical measure, also termed a coefficient of determination, which evaluates how close the data (data points) are to the fitted regression line.

### <font color="red"> 1. Least square method

```python
def least_square_method(x, y):
    #x mean value
    x_mean = sum(x)/len(x)
    #y mean value
    y_mean = sum(y)/len(y)
    #x-x_mean
    x_value = [value - x_mean for value in x]
    #y-y_mean
    y_value = [value - y_mean for value in y]
    #(x-x_mean)*(x-x_mean)
    x_value_square = [x**2 for x in x_value]
    #sum of (x-x_mean)*(x-x_mean)
    x_value_square_total = sum(x_value_square)
    #(x-x_mean)(y-y_mean)
    x_y_value_square = [x*y for x,y in zip(x_value,y_value)]
    #sum of (x-x_mean)(y-y_mean)
    x_y_value_square_total = sum(x_y_value_square)
    #b1 = sigma(x-x_mean)(y-y_mean)/sigma(x-x_mean)*(x-x_mean)
    b1 = x_y_value_square_total/x_value_square_total
    #y = bo + b1 * x
    bo = y_mean - (b1*x_mean)
    #returning and printing to screen
    return x_mean, y_mean, x_value, y_value, x_value_square, x_y_value_square, b1, bo

#calling and storing function into result
result = least_square_method([1,2,3,4,5], [2,4,5,4,5])

#printing function according to suitable result
print(f"1. x mean: {result[0]}")
print(f"2. y mean: {result[1]}")
print(f"3. (x-x_mean): {result[2]}")
print(f"4. (y-y_mean): {result[3]}")
print(f"5. (x - x_mean)^2: {result[4]}")
print(f"6. (x-x_mean)(y-y_mean): {result[5]}")
print(f"7. b1: {result[6]}")
print(f"8. bo: {result[7]}")
print(f"9. y = {result[7]} + {result[6]}x")
```

###  <font color="red">2. R-Squared

#### - R square tells us, how well a regression line predicts or estimates actual values. 

#### - Adjusted R squared 

#### 1. To calculate R squared, we take actual values and we determine the mean of those values. 

#### 2. We look at the distance from the actual value to the mean, so we have distance to the actual value of the mean

#### 3. We draw a regression line  and we comes up with estimated values and these are all estimated values. 


#### 4. we take distance to estimated value of the mean. we compare this with the distance of the actual to the mean. 

#### 5. Defininf function for R Squared and calulating 

```python
bo = 2.2
b1 = 0.6
def r_squared(x,y):
    #0. x mean value
    x_mean = sum(x)/len(x)
    #1. y mean value
    y_mean = sum(y)/len(y)
    #2. (y-y_mean)
    y_value = [value - y_mean for value in y]
    #3. total of y_value_total
    y_value_total = sum(y_value)
    #4. (y-y_mean)*(y-y_mean)
    y_value_square = [y**2 for y in y_value]
    #5. sum of (y-y_mean)*(y-y_mean)
    y_value_square_total = sum(y_value_square)
    #6. estimated value
    y_estimated = [bo + (b1*value) for value in x]
    #7. sum of all y_estimated value
    y_estimated_total = sum(y_estimated)
    #8. (y_estimated - y_mean)
    y_estimated_mins_y_mean = [value - y_mean for value in y_estimated]
    #9. sum of y_estimated_mins_y_mean 
    y_estimated_mins_y_mean_total = sum(y_estimated_mins_y_mean)
    #10. (y_estimated - y_mean)**2
    y_estimated_mins_y_mean_square = [value**2 for value in y_estimated_mins_y_mean]
    #11. sum of (y_estimated - y_mean)**2
    y_estimated_mins_y_mean_square_total = sum(y_estimated_mins_y_mean_square)
    #12. R Square
    r_square = y_estimated_mins_y_mean_square_total/y_value_square_total
    
    #returning and printing result on screen
    return x_mean, y_mean, y_value, y_value_total, y_value_square, y_value_square_total, y_estimated, y_estimated_total,y_estimated_mins_y_mean,y_estimated_mins_y_mean_total,y_estimated_mins_y_mean_square,y_estimated_mins_y_mean_square_total, r_square

#x and y argument for function r_squared
result = r_squared([1,2,3,4,5], [2,4,5,4,5])
print(f"1. x mean: {result[0]}")
print(f"2. y mean: {result[1]}")
print()
print(f"3. (y - y_mean): {result[2]}")
print(f"4. total of (y - y_mean): {result[3]}")
print()
print(f"5. (y-y_mean)(y-y_mean): {result[4]}")
print(f"6. total of (y-y_mean)(y-y_mean): {result[5]}")
print()
print(f"7. y estimated: {result[6]}")
print(f"8. total of y estimated: {result[7]}")
print()
print(f"9. (y_estimated - y_mean): {result[8]}")
print(f"10. total of (y_estimated - y_mean): {result[9]}")
print()
print(f"11. (y_estimated - y_mean)^2:{result[10]}")
print(f"12. total of (y_estimated - y_mean)^2: {result[11]}")
print()
print(f"13. R squared: {round(result[12],2)}")
print()
```

#### Example 3 - Missing value with triple columns

In a multiple linear regression with four independent variables (also known as predictors or features), the goal is to find the best linear relationship between these four variables and the dependent variable (also known as the response or target). This relationship can be represented by a linear equation in the form of

y = b0 + b1x1 + b2x2 + b3x3 + b4x4

1. Where y is the dependent variable, x1, x2, x3, x4 are the independent variables and b0, b1, b2, b3, b4 are the coefficients or parameters of the model.

2. The coefficients are chosen to minimize the sum of the squared residuals (also known as the "least squares" method). This method is used to find the line of best fit which minimizes the difference between the predicted values and the actual values.

3. Once the model is trained, it can be used to make predictions about the dependent variable based on the values of the independent variables.

4. It's important to note that all independent variables should be independent of each other and should not contain multicollinearity, as this can affect the accuracy of the model.

### Example 4 - Multiple missing Values (SimpleImputer Model)

SimpleImputer is a basic imputer which replaces the missing values with a specific value or the mean/median/most_frequent value of the feature.

#### Simple imputation is a technique that is used to replace missing values in a dataset with a chosen value, such as the mean, median, or most frequent value of the feature.

There are several reasons why simple imputation is used for multiple missing values:

1. Ease of implementation: Simple imputation is relatively straightforward to implement and does not require a lot of computational resources. It is a quick and easy way to handle missing values in a dataset.

2. Interpretability: Simple imputation methods such as mean or median are easy to interpret and understand. It is clear what value is being used to replace the missing data.

3. Assumptions: Simple imputation methods make assumptions about the underlying distribution of the data. For example, if the data is normally distributed, then the mean is a good estimate of the central tendency of the data.

4. Effective for small amount of missing data: Simple imputation methods are most effective when the amount of missing data is relatively small. This method is not as effective when there is a large amount of missing data, as it can lead to bias in the estimates.

5. Works well with other techniques: Simple imputation is often used in combination with other techniques such as data visualization, outlier detection, and feature engineering. It can be a useful preprocessing step before applying more advanced techniques.

It's always recommended to check the data distribution before and after imputing and also look at the correlation between the features in case of high correlation you can drop any one of them.

### Example 5 - Multiple missing Values (KNNImputer)

KNNImputer is an imputer which uses the k-Nearest Neighbors algorithm to impute the missing values. It looks at the k nearest rows that have a value for the variable and imputes the missing value with the average value of the k nearest rows.

```python
from sklearn.impute import KNNImputer
```

#### 1. Dataset with missing Values

```python
#dataset with mupltiples missing values
data = {"a":[1,2,3,4,5,6,7,8],
        "b":[1,2,3,4,5,None, None, None], 
        "c":[1,None,None,None,5,6,7,8],
        "d":[1,2,3,4,None,None,7,8], 
        "e":[1,2,3,4,5,6,None, None],
       }
for k,v in data.items():
    print(k,":",v)
```

#### 2. Create a DataFrame from your data

```python
# Create a DataFrame from your data
df = pd.DataFrame(data)
df
```

#### 3. Create an instance of the KNNImputer

```python
imputer = KNNImputer(n_neighbors=27)

# Fit the imputer on the non-missing values
df_non_missing = df.dropna()
imputer.fit(df_non_missing)

# Use the imputer to fill in the missing values
df_filled = imputer.transform(df)

# Create a new DataFrame with the filled values
df_filled = pd.DataFrame(df_filled, columns=df.columns)
```
